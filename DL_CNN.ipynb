```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
data=pd.read_csv("/content/concrete_data.csv")

data.isnull().sum()
data.head()
```

```python
y=data['Strength']
x=data.drop(['Strength'],axis=1)
y.head()
x.head()
```

```python
from sklearn.model_selection import train_test_split

# Randomly split the data into a training and test sets by holding 30% of the data for testing.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
input=x.shape[1]
```

```python
from sklearn.metrics import mean_squared_error

mse_list=[]

for _ in range(50):

    model=Sequential()
    #One hidden layer of 10 nodes, and a ReLU activation function
    model.add(Dense(10,activation="relu",input_shape=(input,)))
    model.add(Dense(1))

    #Use the adam optimizer and the mean squared error  as the loss function.
    model.compile(optimizer="adam",loss="mean_squared_error")

    #Train the model on the training data using 50 epochs.
    model.fit(x_train,y_train,validation_split=0.3, epochs=50, verbose=2)

    #Evaluate the model on the test data and compute the mean squared error
    mse=model.evaluate(x_test,y_test)
    mse_list.append(mse)

mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

print(f'Mean of Mean Squared Errors: {mean_mse}')
print(f'Standard Deviation of Mean Squared Errors: {std_mse}')
```

```python
#Normalize the data
x_teain_norm = (x_train - x_train.mean()) / x_train.std()
x_test_norm = (x_test - x_test.mean()) / x_test.std()

mse_list_norm=[]

for _ in range(50):

    model=Sequential()
    #One hidden layer of 10 nodes, and a ReLU activation function
    model.add(Dense(10,activation="relu",input_shape=(input,)))
    model.add(Dense(1))

    #Use the adam optimizer and the mean squared error  as the loss function.
    model.compile(optimizer="adam",loss="mean_squared_error")

    #Train the model on the training data using 50 epochs.
    model.fit(x_teain_norm,y_train,validation_split=0.3, epochs=50, verbose=2)

    #Evaluate the model on the test data and compute the mean squared error
    mse_norm=model.evaluate(x_test_norm,y_test)
    mse_list_norm.append(mse_norm)

mean_mse_norm = np.mean(mse_list_norm)
std_mse_norm = np.std(mse_list_norm)

print(f'Mean of Mean Squared Errors: {mean_mse_norm}')
print(f'Standard Deviation of Mean Squared Errors: {std_mse_norm}')
```

```python
#Increate the number of epochs
mse_list2=[]
for _ in range(50):

    model2=Sequential()
    #One hidden layer of 10 nodes, and a ReLU activation function
    model2.add(Dense(10,activation="relu",input_shape=(input,)))
    model2.add(Dense(1))

    #Use the adam optimizer and the mean squared error  as the loss function.
    model2.compile(optimizer="adam",loss="mean_squared_error")

    #Train the model on the training data using 50 epochs.
    model2.fit(x_teain_norm,y_train,validation_split=0.3, epochs=100, verbose=2)

    #Evaluate the model on the test data and compute the mean squared error
    mse2=model2.evaluate(x_test_norm,y_test)
    mse_list2.append(mse2)

mean_mse2 = np.mean(mse_list2)
std_mse2 = np.std(mse_list2)

print(f'Mean of Mean Squared Errors: {mean_mse2}')
print(f'Standard Deviation of Mean Squared Errors: {std_mse2}')
```

```python
# Increase the number of hidden layers
#Three hidden layers, each of 10 nodes and ReLU activation function
mse_list3=[]

for _ in range(50):
    model=Sequential()
    #One hidden layer of 10 nodes, and a ReLU activation function
    model.add(Dense(10,activation="relu",input_shape=(input,)))
    model.add(Dense(10,activation="relu"))
    model.add(Dense(10,activation="relu"))
    model.add(Dense(1))

    #Use the adam optimizer and the mean squared error  as the loss function.
    model.compile(optimizer="adam",loss="mean_squared_error")

    #Train the model on the training data using 50 epochs.
    model.fit(x_teain_norm,y_train,validation_split=0.3, epochs=100, verbose=2)

    #Evaluate the model on the test data and compute the mean squared error
    mse3=model.evaluate(x_test_norm,y_test)
    mse_list3.append(mse3)

mean_mse3 = np.mean(mse_list3)
std_mse3 = np.std(mse_list3)

print(f'Mean of Mean Squared Errors: {mean_mse3}')
print(f'Standard Deviation of Mean Squared Errors: {std_mse3}')
```
